## Linear regression model for multiple features using vectorization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
Data=pd.read_csv('Real_estate.csv').set_index("No")
```

```python
Data.rename(columns={'X1 transaction date':'X1','X2 house age':'X2','X3 distance to the nearest MRT station':'X3',
                     'X4 number of convenience stores':'X4','X5 latitude':'X5','X6 longitude':'X6','Y house price of unit area':'Y'},inplace=True)
m=len(Data)
```

```python
X=(np.array([np.ones(m),Data['X1'],Data['X2'],Data['X3'],Data['X4'],Data['X5'],Data['X6']])).T
```

```python
Y=np.array(Data['Y'])
Y=Y.reshape(m,1)
Y0=Y.reshape(414)
```

```python
Q=np.linalg.inv((X.T)@X)@((X.T)@Y)
Q=Q.reshape(7)
```

```python
def L_eqn(X):
    X@Q
```

## Errors compared to real values real values

```python
for i in range(414):
    plt.plot(np.arange(0,414),(X@Q)-Y0)
```
